# Adaptive thresholding - Gaussian.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title, cmap=None):
    if cmap == None:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap=cmap)
    
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = cv2.imread('images/global_thresh_problem.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
show_image(image, 'Global thresholding problem.')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show_image(gray_image, 'Global thresholding problem - Grayscale', cmap='gray')
```

```python
adaptive_gauss = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 9)
```

```python
show_image(adaptive_gauss, 'Adaptive thresholding - Gaussian', 'gray')
```

As we can see, we get a much better result than previously.
