# Canny Edge Detection - Coins Challenge

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/coins.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

title = 'Coins Challenge'

plt.imshow(image)
plt.title(title)
plt.axis('off')

plt.show()
```

```python
gray_image  = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title(f'{title} - Grayscale')
plt.axis('off')

plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5),  0)

plt.imshow(blurred_image, cmap='gray')
plt.title(f'{title} - Blurred')
plt.axis('off')

plt.show()
```

```python
canny_image = cv2.Canny(blurred_image, 80, 140)

plt.imshow(canny_image, cmap='gray')
plt.title(f'{title} - Canny Edge')
plt.axis('off')

plt.show()
```
