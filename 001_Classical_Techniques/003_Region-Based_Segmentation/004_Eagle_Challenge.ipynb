# Region-Based - Eagle Challenge

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/eagle.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title('Eagle')
plt.axis('off')

plt.show()
```

```python
HEIGHT, WIDTH = image.shape[0], image.shape[1]

print(f'Image height: {HEIGHT} px')
print(f'Image width: {WIDTH} px')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title('Eagle - Grayscale')
plt.axis('off')

plt.show()
```

```python
flattened_image = gray_image.flatten()

print(f'Flattened image: {flattened_image.shape}')
```

## Example 01 - Two regions:

```python
normalized_flattened_image = flattened_image.copy() / 255
mean_pixels = normalized_flattened_image.mean()
number_pixels = len(normalized_flattened_image)

print(f'Mean of pixels value: {mean_pixels}')
print(f'Number of pixels: {number_pixels}')
```

```python
for i in range(number_pixels):
    if normalized_flattened_image[i] > mean_pixels:
        normalized_flattened_image[i] = 1
    else:
        normalized_flattened_image[i] = 0
```

```python
region_based_image = normalized_flattened_image.reshape(HEIGHT, WIDTH)

plt.imshow(region_based_image, cmap='gray')
plt.title('Eagle - Region-Based - Example 01')
plt.axis('off')

plt.show()
```

## Example 02 - Three regions:

Given the interval of pixel values between 0 and 1, we want to get three regions of the image. For this, we divide 1 by 3, which results in 0.33, which is the magnitude of each region.

<i>1 Region:</i>

The pixel values are between 0 and 0.33.

<i>2 Region:</i>

The pixel values are between 0.34 and 0.66.

<i>3 Region:</i>

The pixel values are between 0.67 and 1.

```python
normalized_flattened_image = flattened_image.copy() / 255
number_pixels = len(normalized_flattened_image)

print(f'Number of pixels: {number_pixels}')
```

```python
for i in range(number_pixels):
    if normalized_flattened_image[i] <= 0.33:
        normalized_flattened_image[i] = 0
    elif normalized_flattened_image[i] > 0.33 and normalized_flattened_image[i] <= 0.66:
        normalized_flattened_image[i] = 0.5
    else:
        normalized_flattened_image[i] = 1
```

```python
region_based_image = normalized_flattened_image.reshape(HEIGHT, WIDTH)

plt.imshow(region_based_image, cmap='gray')
plt.title('Eagle - Region-Based - Example 02')
plt.axis('off')

plt.show()
```
