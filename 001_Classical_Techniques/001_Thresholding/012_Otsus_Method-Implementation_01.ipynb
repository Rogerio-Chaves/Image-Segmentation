# Otsu's method Implementation.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title, cmap=None):
    if cmap == None:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap=cmap)
    
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = cv2.imread('images/mountain.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
show_image(image, 'Mountain')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show_image(gray_image, 'Mountain - Grayscale', cmap='gray')
```

```python
gray_image.shape
```

```python
threshold_value, thresholded_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
print(f'Thresholded value by Otsu\'s Method: {threshold_value}')
```

```python
plt.imshow(thresholded_image)
show_image(thresholded_image, 'Mountain - Thresholded', cmap='gray')
```

Getting the histogram.

```python
image_flattened = gray_image.flatten()

plt.hist(x=image_flattened, bins=255)
plt.title('Histogram - Frequency of each pixel value in the image.')
plt.text(x=10, y=2500, s='Region 1')
plt.text(x=80, y=5000, s='Region 2')
plt.text(x=165, y=17500, s='Region 3')
plt.vlines(threshold_value, 0, 20000, 'black')
plt.show()
```
