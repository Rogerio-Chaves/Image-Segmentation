
# Thermal Challenge 1.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
def show_image(image, title, cmap=None):
    figure = plt.gcf()
    figure.set_size_inches(18,6)
    if cmap == None:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap=cmap)
    
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('images/thermal_01.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
show_image(image, 'Thermal 01')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show_image(gray_image, 'Thermal 01 - Grayscale', cmap='gray')
```

```python
threshold = 175
```

```python
titles = ['Original', 'Binary thresholding', 'Inverted binary thresholding', 
    'Truncated thresholding', 'To zero thresholding', 
    'To inverted zero thresholding']



figure = plt.figure(figsize=(10,6))
axes = figure.subplots(2,3)
figure.suptitle('Thermal 01 images')



axes[0,0].imshow(gray_image, cmap='gray')
axes[0,0].set_title(titles[0])
axes[0,0].axis('off')

_, thresh_binary = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_BINARY)
axes[0,1].imshow(thresh_binary, cmap='gray')
axes[0,1].set_title(titles[1])
axes[0,1].axis('off')

_, thresh_binary_invert = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_BINARY_INV)
axes[0,2].imshow(thresh_binary_invert, cmap='gray')
axes[0,2].set_title(titles[2])
axes[0,2].axis('off')

_, thresh_trunc = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_TRUNC)
axes[1,0].imshow(thresh_trunc, cmap='gray')
axes[1,0].set_title(titles[3])
axes[1,0].axis('off')

_, thresh_tozero = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_TOZERO)
axes[1,1].imshow(thresh_tozero, cmap='gray')
axes[1,1].set_title(titles[4])
axes[1,1].axis('off')

_, thresh_tozero_invert = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_TOZERO_INV)
axes[1,2].imshow(thresh_tozero_invert, cmap='gray')
axes[1,2].set_title(titles[5])
axes[1,2].axis('off')

plt.show()
```
