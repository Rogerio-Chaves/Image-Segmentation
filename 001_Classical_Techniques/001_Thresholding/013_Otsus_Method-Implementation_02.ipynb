# Otsu's method Implementation - Image with noise

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title, cmap=None):
    if cmap == None:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap=cmap)
    
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = cv2.imread('images/sheet_with_noise.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
show_image(image, 'Sheet')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show_image(gray_image, 'Sheet - Gray Scale', cmap='gray')
```

```python
threshold_value, thresholded_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
print(f'Thresholded value by Otsu\'s Method: {threshold_value}')
```

```python
plt.imshow(thresholded_image)
show_image(thresholded_image, 'Sheet - Thresholded', cmap='gray')
```

Getting the histogram.

```python
image_flattened = gray_image.flatten()

plt.hist(x=image_flattened, bins=255)
plt.title('Histogram - Frequency of each pixel value in the image.')
plt.vlines(threshold_value, 0, np.unique(image_flattened, return_counts=True)[1].max(), 'black')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5,5), 0)
```

```python
show_image(blurred_image, 'Sheet - Bluerred', 'gray')
```

```python
threshold_blurred_image, thresholded_blurred_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
print(f'Thresholded value by Otsu\'s Method: {threshold_blurred_image}')
```

```python
plt.imshow(thresholded_blurred_image)
show_image(thresholded_blurred_image, 'Sheet - Thresholded', cmap='gray')
```
