# Truncated Thresholding.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

```python
def show_image(image, title, cmap=None):
    figure = plt.gcf()
    figure.set_size_inches(18,6)
    if cmap == None:
        plt.imshow(image)
    else:
        plt.imshow(image, cmap=cmap)
    
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('images/landscape.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
show_image(image, 'Landscape')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show_image(gray_image, 'Landscape - Grayscale', cmap='gray')
```

```python
threshold = 140
```

```python
_, thresholded_image = cv2.threshold(gray_image, threshold, 255, cv2.THRESH_TRUNC)
```

```python
show_image(thresholded_image, 'Landscape - Truncated thresholding', 'gray')
```
