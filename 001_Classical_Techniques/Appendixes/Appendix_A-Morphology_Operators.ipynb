# Apendice A - Morphology Operators


The type of morphology operators are:

- Erosion;

- Dilation;

- Opening;

- Closing.

## Erosion

Erosion is one of the basic operators in the area of mathematical morphology, and it is typically applied to binary images. The basic effect is to erode the boundaries of regions of the foreground (the white region). In this operation, areas of foreground pixels shrink in size, and holes within those areas become larger [1].

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
thresholded_image = cv2.imread('landscape-inverted_binary_thresholded.png')
thresholded_image = cv2.cvtColor(thresholded_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(thresholded_image)
plt.title('Landscape - Threhsolded')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
eroded_image = cv2.erode(thresholded_image, kernel)
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(eroded_image)
plt.title('Landscape - Eroded')
plt.axis('off')

plt.show()
```

## Dilation

Dilation is one of the basic operators in the area of mathematical morphology, and it is typically applied to binary images. The basic effect is to gradually enlarge the boundaries of regions of foreground pixels (the white region). In this operation, the area of foreground pixels grows in size while other areas become smaller [2].

```python
kernel = np.ones((3, 3), np.uint8)
dilated_image = cv2.dilate(thresholded_image, kernel)
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_image)
plt.title('Landscape - Dilated')
plt.axis('off')

plt.show()
```

## Opening

The opening operator is derived from erosion and dilation operators. Like these operators, opening is an operator applied to binary images. This operator is defined as follows: first the erosion operator is applied, and then the dilation operator is applied. The effect basic of this operator is like the erode opertor, but less destructive [3].

 Opening = Erosion + Dilation
 
It is used to remove small objects from an image while preserving the
shape and size of larger objects in the image [4].

```python
kernel = np.ones((3, 3), np.uint8)
eroded_image = cv2.erode(thresholded_image, kernel)
dilated_image = cv2.dilate(eroded_image, kernel)

opened_image = dilated_image
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(opened_image)
plt.title('Landscape - Opened')
plt.axis('off')

plt.show()
```

## Closing

The closing operator is derived from erosion and dilation operators, similar with the opening operator. In the closing operator, first the dilation operator is applied, and then the erosion operator is applied [5].

Closing = Dilation + Erosion

This operator is usually used to remove discontinuities in regions of the image [4].

```python
kernel = np.ones((3, 3), np.uint8)
dilated_image = cv2.dilate(thresholded_image, kernel)
eroded_image = cv2.erode(dilated_image, kernel)

closed_image = eroded_image
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(closed_image)
plt.title('Landscape - Closed')
plt.axis('off')

plt.show()
```
