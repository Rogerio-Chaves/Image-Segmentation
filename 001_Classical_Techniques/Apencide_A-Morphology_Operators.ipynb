# Apendice A - Morphology Operators


The type of morphology operators are:

- Erosion;

- Dilation;

- Opening;

- Closing.

## Erosion

Erosion is one of the basic operators in the area of mathematical morphology, and it is typically applied to binary images. The basic effect is to erode the boundaries of regions of the foreground (the white region). In this operation, areas of foreground pixels shrink in size, and holes within those areas become larger [1].

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
thresholded_image = cv2.imread('landscape-inverted_binary_thresholded.png')
thresholded_image = cv2.cvtColor(thresholded_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(thresholded_image)
plt.title('Landscape - Threhsolded')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
eroded_image = cv2.erode(thresholded_image, kernel)

plt.imshow(eroded_image)
plt.title('Landscape - Eroded')
plt.axis('off')
plt.show()
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(eroded_image)
plt.title('Landscape - Eroded')
plt.axis('off')

plt.show()
```

## Dilation

Dilation is one of the basic operators in the area of mathematical morphology, and it is typically applied to binary images. The basic effect is to gradually enlarge the boundaries of regions of foreground pixels (the white region). In this operation, the area of foreground pixels grows in size while other areas become smaller [2].

```python
kernel = np.ones((3, 3), np.uint8)
dilated_image = cv2.dilate(thresholded_image, kernel)

plt.imshow(dilated_image)
plt.title('Landscape - Dilated')
plt.axis('off')
plt.show()
```

```python
figure = plt.gcf()
figure.set_size_inches(10, 7)

plt.subplot(1, 2, 1)
plt.imshow(thresholded_image)
plt.title('Landscape - Thresholded')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_image)
plt.title('Landscape - Dilated')
plt.axis('off')

plt.show()
```
